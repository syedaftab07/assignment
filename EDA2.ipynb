```python
df1=df.copy()
df2=df.copy()

from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
df1[['age','education_num','capital_gain','capital_loss','hours_per_week']] = min_max_scaler.fit_transform(df[['age','education_num','capital_gain','capital_loss','hours_per_week']])
df1.describe()

```

**Min Max**
It is used when you want all your features with different unit to be of same unit and when you data does not have  any outliers.

```python
standard_scaler = preprocessing.StandardScaler()
df2[['age','education_num','capital_gain','capital_loss','hours_per_week']] = standard_scaler.fit_transform(df[['age','education_num','capital_gain','capital_loss','hours_per_week']])
df2.describe()
```

**Standard Scaler**

It is used when you data follows gaussian distribution or your algorithm works better with mean of 0 and standard deviation one ,
For exmaple: Linear Regression, Logestic regression

```python
from sklearn.preprocessing import OneHotEncoder
# One-Hot Encoding for categorical variables with less than 5 categories
one_hot_cols = [col for col in df.columns if df[col].nunique() < 5 and df[col].dtype == 'object']
one_hot_encoder = OneHotEncoder(drop='first', sparse=False)
one_hot_encoded = pd.DataFrame(one_hot_encoder.fit_transform(df[one_hot_cols]), columns=one_hot_encoder.get_feature_names_out(one_hot_cols))
df.drop(one_hot_cols, axis=1, inplace=True)
df = pd.concat([df, one_hot_encoded], axis=1)
```

This technique is suitable for variables with few categories as it avoids creating redundant ordinal relationships between categories.


```python
from sklearn.preprocessing import LabelEncoder
# Label Encoding for categorical variables with more than 5 categories
label_cols = [col for col in df.columns if df[col].nunique() >= 5 and df[col].dtype == 'object']
label_encoder = LabelEncoder()
for col in label_cols:
    df[col] = label_encoder.fit_transform(df[col])
```

Label Encoding is appropriate when the categorical variable has many categories, as it preserves the ordinal relationship between categories.

```python
# Create New Features
df['work_hours_ratio'] = df['hours_per_week'] / df['education_num']
```

We created a new feature called 'work_hours_ratio', representing the ratio of hours_per_week to education_num. This feature captures the proportion of weekly working hours relative to the level of education.

```python
# Apply Transformation
df['capital_gain_log'] = np.log1p(df['capital_gain'])
print('Capital gain',df['capital_gain'].skew())
print('Capital gain log',df['capital_gain_log'].skew())
```

We applied a log transformation to the 'capital_gain' feature to reduce skewness and make the distribution more symmetric. Log transformation is beneficial for features with heavily skewed distributions, as it helps to stabilize variance and make the data more interpretable

```python
from sklearn.ensemble import IsolationForest
import ppscore as pps
# Outlier Detection using Isolation Forest
iso_forest = IsolationForest(contamination=0.05, random_state=42)
outliers = iso_forest.fit_predict(df)
df = df[outliers == 1]
```

We used the Isolation Forest algorithm to identify and remove outliers from the dataset.Removing outliers can lead to more robust and accurate models.

```python
# Apply PPS to find feature relationships
pps_matrix = pps.matrix(df)

print("\nPPS Matrix:")
print(pps_matrix)
```

PPS provides insights into both linear and non-linear relationships between features, offering a more comprehensive view of feature interactions compared to traditional correlation analysis.
